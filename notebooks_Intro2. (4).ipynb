```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

# Generate a regression dataset similar to California housing
X, y = make_regression(n_samples=1000, n_features=8, noise=0.1, random_state=42)

# Print shapes
print("X.shape:", X.shape)
print("y.shape:", y.shape)

# Optional: visualize first two features against target
plt.scatter(X[:, 0], y, alpha=0.5)
plt.xlabel("Feature 0")
plt.ylabel("Target")
plt.title("Regression dataset similar to California Housing")
plt.show()
```
