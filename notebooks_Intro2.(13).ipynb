```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Create the wave dataset (like mglearn's wave)
np.random.seed(42)
X = np.linspace(-3, 3, 40).reshape(-1, 1)           # 40 points between -3 and 3
y = np.sin(X).ravel() + np.random.normal(0, 0.3, X.shape[0])  # sine wave + noise

# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Create and train linear regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict values for a smooth curve
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_plot = lr.predict(X_plot)

# Plot the linear regression on the wave dataset
plt.figure(figsize=(8, 6))
plt.plot(X_plot, y_plot, label="Linear regression", color='green')
plt.scatter(X_train, y_train, marker='^', color='blue', s=80, label="Training data")
plt.scatter(X_test, y_test, marker='v', color='red', s=80, label="Test data")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression on Wave Dataset")
plt.legend()
plt.show()

# Print R^2 scores
print("Training set R^2: {:.2f}".format(lr.score(X_train, y_train)))
print("Test set R^2: {:.2f}".format(lr.score(X_test, y_test)))
```
