```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.datasets import make_blobs

# Create dataset (3 classes)
X, y = make_blobs(centers=3, random_state=42)

# Train model
linear_svm = LinearSVC(max_iter=10000)
linear_svm.fit(X, y)

# Print shapes
print("Coefficient shape:", linear_svm.coef_.shape)
print("Intercept shape:", linear_svm.intercept_.shape)

# Plot data points (replace mglearn.discrete_scatter)
for i in range(3):
    plt.scatter(X[y == i, 0], X[y == i, 1], label=f"Class {i}")

# Plot decision boundaries
line = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 100)

for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, ['b', 'r', 'g']):
    plt.plot(line, -(coef[0] * line + intercept) / coef[1], c=color)

# Labels and legend
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend()
plt.show()
```
