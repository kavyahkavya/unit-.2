```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression

# Create a simple 1D regression dataset
X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Create KNN regressors with 1 and 3 neighbors
knn1 = KNeighborsRegressor(n_neighbors=1)
knn3 = KNeighborsRegressor(n_neighbors=3)

# Train models
knn1.fit(X_train, y_train)
knn3.fit(X_train, y_train)

# Plot predictions
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_plot_1 = knn1.predict(X_plot)
y_plot_3 = knn3.predict(X_plot)

plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, c='blue', label='Training data')
plt.plot(X_plot, y_plot_1, color='red', label='KNN regression (k=1)')
plt.plot(X_plot, y_plot_3, color='green', label='KNN regression (k=3)')
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("KNN Regression with k=1 and k=3")
plt.legend()
plt.show()
```
